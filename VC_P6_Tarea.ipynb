```python
import os
import cv2
from deepface import DeepFace
import FaceNormalizationUtils as faceutils
import FaceDetectors
import imutils
```

# Filtro con emoticono en función de tu estado de ánimo

```python
smiley_img = cv2.imread('risa.png')
sad_img = cv2.imread('triste.png')
neutral_img = cv2.imread('neutral.png')
sorpresa_img = cv2.imread('sorpresa.png')
enfado_img = cv2.imread('enfado.png')

FDet = FaceDetectors.FaceDetector()

vid = cv2.VideoCapture(0)

font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
color = (255, 0, 0)
thickness = 2
pos = (50, 50)

while True:

    ret, frame = vid.read()

    
    values = FDet.SingleFaceEyesDetection(frame, FDet.FaceDetectors[1], FDet.EyeDetectors[1])

    if ret and (values is not None):

        face, eyes, shape = values
        x, y, w, h = face

        obj = DeepFace.analyze(img_path= frame, enforce_detection=False, actions=['gender', 'emotion'])
        print(obj)

        image = cv2.putText(frame, str(obj[0]['dominant_gender']+ ' ' + obj[0]['dominant_emotion']), pos, font, fontScale, color, thickness, cv2.LINE_AA)

        try:
            # Intenta redimensionar la imagen feliz
            if obj[0]['dominant_emotion'] == 'happy':
                smiley_resized = imutils.resize(smiley_img, width=w)
                frame[y:y + smiley_resized.shape[0], x:x + smiley_resized.shape[1]] = smiley_resized

            # Intenta redimensionar la imagen enfado
            elif obj[0]['dominant_emotion'] == 'angry':
                miedo_resized = imutils.resize(enfado_img, width=w)
                frame[y:y + miedo_resized.shape[0], x:x + miedo_resized.shape[1]] = miedo_resized

            # Intenta redimensionar la imagen triste
            elif obj[0]['dominant_emotion'] == 'sad':
                sad_resized = imutils.resize(sad_img, width=w)
                frame[y:y + sad_resized.shape[0], x:x + sad_resized.shape[1]] = sad_resized

            # Intenta redimensionar la imagen neutral
            elif obj[0]['dominant_emotion'] == 'neutral':
                neutral_resized = imutils.resize(neutral_img, width=w)
                frame[y:y + neutral_resized.shape[0], x:x + neutral_resized.shape[1]] = neutral_resized

            # Intenta redimensionar la imagen sorpresa
            elif obj[0]['dominant_emotion'] == 'surprise':
                sorprise_resized = imutils.resize(sorpresa_img, width=w)
                frame[y:y + sorprise_resized.shape[0], x:x + sorprise_resized.shape[1]] = sorprise_resized

        except Exception as e:
            print(f"Error al procesar la imagen: {e}")


        cv2.imshow('Vid', frame)

    if cv2.waitKey(20) == 27:
        break

vid.release()
cv2.destroyAllWindows()
```

# Barras que indican el estado de ánimo

```python
vid = cv2.VideoCapture(0)

# font 
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
color = (255, 0, 0) 
thickness = 2 
pos = (50, 50) 
  
ret, frame = vid.read()
height, width = frame.shape[:2]
surprise = 20
fear = 20
happy = 20
angry = 20
sad = 20

while(True):

    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:  
        
        obj = DeepFace.analyze(img_path = frame, enforce_detection=False, actions =['gender', 'emotion'])
        print(obj)

        # Primera cara
        image = cv2.putText(frame, str(obj[0]['dominant_gender']+' '+obj[0]['dominant_emotion']), pos, font,  
                        fontScale, color, thickness, cv2.LINE_AA) 

        

        if str(obj[0]['dominant_emotion']) == 'surprise' and surprise < width-20:
            surprise += (width-40)//15
        elif surprise > 20:
            surprise -= 5

        if str(obj[0]['dominant_emotion']) == 'fear' and fear < width-20:
            fear += (width-40)//15
        elif fear > 20:
            fear -= 5     

        if str(obj[0]['dominant_emotion']) == 'happy' and happy < width-20:
            happy += (width-40)//15
        elif happy > 20:
            happy -= 5      

        if str(obj[0]['dominant_emotion']) == 'angry' and angry < width-20:
            angry += (width-40)//15
        elif angry > 20:
            angry -= 5     

        if str(obj[0]['dominant_emotion']) == 'sad' and sad < width-20:
            sad += (width-40)//15
        elif sad > 20:
            sad -= 5
        

        
        # surprise bar
        cv2.putText(frame, "Surprise", (width-100,30), font, 0.75, (255, 255, 255), 2)
        cv2.rectangle(frame, (20, height-100), (surprise, height-90), (255, 255, 255), -1)        
        
        # Fear bar
        cv2.putText(frame, "Fear", (width-100,60), font, 0.75, (58, 35, 62), 2)
        cv2.rectangle(frame, (20, height-80), (fear, height-70), (58, 35, 62), -1)
        
        # Happy bar
        cv2.putText(frame, "Happy", (width-100,90), font, 0.75, (0,255,255), 2)
        cv2.rectangle(frame, (20, height-60), (happy, height-50), (0, 255, 255), -1)

        # Angry bar
        cv2.putText(frame, "Angry", (width-100,120), font, 0.75, (0,0,255), 2)
        cv2.rectangle(frame, (20, height-40), (angry, height-30), (0, 0, 255), -1) 

        # Sad bar
        cv2.putText(frame, "Sad", (width-100,150), font, 0.75, (255,255,0), 2)
        cv2.rectangle(frame, (20, height-20), (sad, height-10), (255, 255, 0), -1)

        # Muestra fotograma
        cv2.imshow('Vid', frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
